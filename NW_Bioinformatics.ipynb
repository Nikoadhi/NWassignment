```
import numpy as np
import pandas as pd

#initialize sequence
seq1 = 'MHHNALQRRTVWVNAY'
seq2 = 'MHHALQRRTVWVNAY'

mat = np.zeros((len(seq1)+1, len(seq2)+1))

pd.DataFrame(mat)
```

```
#make dotplot matrix 
d_mat = np.zeros((len(seq1), len(seq2)))
for i in range(len(seq1)):
    for j in range(len(seq2)):
        if seq1[i] == seq2[j]:
            d_mat[i][j] = 1
        else:     
            d_mat[i][j] = 0

d_mat
pd.DataFrame(d_mat)
```

```
match_score = 1
nomatch_score = 0
gap_penalty = -2
```

```
def NW(mat):
   for i in range(len(seq1)+1):
       mat[i][0] = i*gap_penalty
   for j in range(len(seq2)+1):
       mat[0][j] = j*gap_penalty
#comparing values
   for i in range(1,len(seq1)+1):
       for j in range(1,len(seq2)+1):
                mat[i][j] = max(mat[i-1][j-1]+d_mat[i-1][j-1], mat[i-1][j]+gap_penalty, mat[i][j-1]+gap_penalty)

   return(mat)

NW(mat)
pd.DataFrame(mat)
```

```
ResultA = ""
ResultB = ""
i = len(seq1)
j = len(seq2)
while (i>0 or j>0):

  if (i>0 and j>0 and mat[i][j] == mat[i-1][j-1] + d_mat[i-1][j-1]):
    ResultA = seq1[i-1] + ResultA
    ResultB = seq2[j-1] + ResultB
    i = i - 1
    j = j - 1

  elif (i>0 and mat[i][j] == mat[i-1][j] + gap_penalty):
    ResultA = seq1[i-1] + ResultA
    ResultB = "-" + ResultB
    i = i - 1
  
  else:
    ResultA = "-" + ResultA
    ResultB = seq2[j-1] + ResultB
    j = j - 1
```
